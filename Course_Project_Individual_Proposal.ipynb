Analysis 1: Relationship Between Introversion and Relationship Satisfaction

Research Question

Is there a significant difference in relationship satisfaction between individuals who prefer solitude (introverted) and those who do not, and how does their desired number of close relationships relate to this satisfaction?

Variables

    Introversion Indicators:
        PSYCH_need_to_belong_scale_dislike_being_alone: Agreement with "I do not like being alone." (Lower scores indicate preference for being alone)
        PSYCH_attachment_style_questionnaire_feel_left_out_or_alone: Agreement with "I often feel left out or alone." (Lower scores may indicate comfort with being alone)

    Desired Number of Close Relationships:
        CONNECTION_social_num_close_relationships_preference: "How many close relationships would you like to have?"

    Relationship Satisfaction Measures:
        WELLNESS_satisfied_relationship: General satisfaction with romantic relationships
        CONNECTION_satisfied_num_friends: Satisfaction with the number of friends
        WELLNESS_life_satisfaction: Overall life satisfaction on a scale of 1 to 10

Visualizations

    Histograms of introversion scores to understand their distribution.
    Box and whisker plots comparing relationship satisfaction between introverted and non-introverted groups.
    Scatter plots showing the relationship between introversion scores and desired number of close relationships.

Analysis Method

    Grouping: Classify participants into 'Introverted' and 'Non-Introverted' based on their scores on introversion indicators (using an educated threshold such as splitting at median. This will of course, be recorded as an assumption).

    Hypotheses:
        Null Hypothesis (H₀): There is no significant difference in relationship satisfaction between introverted and non-introverted individuals.
        Alternative Hypothesis (H₁): Introverted individuals have different levels of relationship satisfaction compared to non-introverted individuals.

    Statistical Tests:
        Bootstrap Sampling: Estimate the sampling distribution of the mean difference in relationship satisfaction between groups.
        Confidence Intervals: Construct confidence intervals for the difference in means.
        P-values: Calculate to determine the statistical significance of observed differences.
        Pearson Correlation: Assess the relationship between introversion scores and the desired number of close relationships.

    Assumptions to Address:
        Independence of Observations: We check if each participant's responses are independent.
        Linearity: For the correlation analysis, we may want to ensure a linear relationship between variables.
        Normality: Using the bootstrap method help mitigate normality concerns.

Relevance to Research Question

This analysis can reveal whether a preference for solitude (being introverted) affects individuals' satisfaction with their relationships and their desire for social connections. This provides insight into the social needs of introverted individuals.

Analysis 2: Impact of Marital Status on Relationship Satisfaction and Number of Close Friends

Research Question

Do married individuals report higher relationship satisfaction and have a different number of close friends compared to those who are single, dating, or in a non-marital relationship?

Variables

    Marital Status:
        DEMO_relationship_status: "What is your current relationship status?" (Categories: Single, Dating, In a relationship, Married)

    Satisfaction Measures:
        WELLNESS_satisfied_relationship: Satisfaction with romantic relationships
        WELLNESS_life_satisfaction: Overall life satisfaction
        ORIGINAL_social_life_evaluations_less_connected_than_preferred: Agreement with "I am less socially connected than I would like to be"

    Number of Close Friends:
        CONNECTION_social_num_close_friends_grouped: "How many close friends do you have?"

Visualizations

    Box and whisker plots of relationship satisfaction by marital status.
    Bar charts showing the average number of close friends across marital status groups.
    Histograms of close friends within each marital status category.

Analysis Method

    Hypotheses:
        Null Hypothesis (H₀): There is no significant difference in relationship satisfaction and number of close friends among different marital status groups.
        Alternative Hypothesis (H₁): Married individuals have higher relationship satisfaction and a different number of close friends compared to other groups.

    Statistical Tests:
        Bootstrap Sampling: Estimate the sampling distribution of means for each group.
        Confidence Intervals: Construct for the differences between group means.
        P-values: Calculate to assess the significance of observed differences.

    Assumptions to Address:
        Independence of Observations: We want to ensure responses are independent across participants.
        Sampling Distribution: We use bootstrap methods to accommodate non-normal distributions since we do not know the underlying distribution.

Relevance to Research Question

Understanding how marital status influences relationship satisfaction and social connections can inform programs aimed at enhancing well-being across different relationship stages.

Analysis 3: Effect of Work-Life Prioritization and Relationship Status on Social Satisfaction

Revised Research Question

Is there a significant relationship between prioritizing work life, relationship status, and satisfaction with social connections and relationships?

Variables

    Work-Life Prioritization Indicators:
        LIFESTYLE_time_use_priorities_work_actual: Priority level assigned to work in actual daily life (1 = top priority, 13 = lowest priority)
        WORK_hours_per_week: Average working hours per week
        LIFESTYLE_time_use_balance_working: Perception of time spent working (Too much, Too little, Just right)

    Relationship Status:
        DEMO_relationship_status: "What is your current relationship status?"

    Social Satisfaction Measures:
        WELLNESS_satisfied_relationship: Satisfaction with romantic relationships
        FRIENDSHIP_hall_friendship_standards_self_social_connections: Agreement with "I have social connections."
        WELLNESS_life_satisfaction: Overall life satisfaction

Suggested Visualizations

    Scatter plots showing the relationship between work priority and social satisfaction scores.
    Box and whisker plots comparing social satisfaction across different relationship statuses and work priorities.
    Histograms of working hours to visualize their distribution.

Analysis Method

    Hypotheses:
        Null Hypothesis (H₀): There is no significant relationship between work prioritization, relationship status, and social satisfaction.
        Alternative Hypothesis (H₁): Higher work prioritization is associated with lower social satisfaction, potentially influenced by relationship status.

    Statistical Tests:
        Pearson Correlation: Assess the relationship between work priority levels (and working hours) and social satisfaction measures.
        Simple Linear Regression: Predict social satisfaction based on work prioritization.
            Model: Social Satisfaction = β₀ + β₁*(Work Priority) + ε
        Bootstrap Sampling: Estimate confidence intervals for correlation coefficients and regression parameters.
        P-values: Determine the significance of correlations and regression coefficients.

    Assumptions to Address:
        Linearity: We need to verify linear relationships between independent and dependent variables.
        Independence of Observations: We need to check if each participant's data is independent.
        Normality of Residuals: We check normality in regression residuals (bootstrap methods can help if violated).

Relevance to Research Question

This analysis will highlight whether prioritizing work affects social satisfaction and how relationship status may interact with this relationship. This provides insights for work-life balance initiatives.
