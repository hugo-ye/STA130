```python
#Q3
# The code suggested by chatgpt with my prompt: 
# Suggest a simpler, more streamlined way 
# to code up this for loop simulation 
# so the process is more clear and easier to understand

#imports the numpy library for using random choice
import numpy as np

# Define the three doors in the game in the form of a python list
all_doors = [1, 2, 3]
# Declare the initial choice to be door 1
initial_choice = 1
# Initialize a counter to track the number of times the player wins
wins = 0

# The number of repetitions of running the simulation
reps = 100000

#Loop through the number of repetitions the simulation 
for _ in range(reps):
    # Step 1: Randomly choose the winning door
    winning_door = np.random.choice(all_doors)

    # Step 2: Reveal a goat door (any door that is not the winning door or the player's initial choice)
    # Creates a list of doors that are not the winning door or the initial choice
    remaining_doors = [door for door in all_doors if door != winning_door and door != initial_choice]
    # Randomly pick a door from the remaining doors to reveal a goat door
    revealed_goat = np.random.choice(remaining_doors)

    # Step 3: Switch to the other remaining door
    # Creates a list that has the door that is neither the initial choice nor the revealed goat
    switched_choice = [door for door in all_doors if door != initial_choice and door != revealed_goat][0]

    # Step 4: Check if switching leads to a win
    if switched_choice == winning_door:
        # Increments the win counter if the switched door matches the winning door
        wins += 1

# Calculate and print the win percentage
win_rate = wins / reps
print(f"Win rate by switching: {win_rate:.2%}")
```
